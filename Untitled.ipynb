```python
df = df.drop(['Unnamed: 0', 'total_votes20', 'votes20_Donald_Trump','votes20_Joe_Biden','IncomeErr','IncomePerCapErr',
        'Carpool', 'Transit','MeanCommute', 'PrivateWork'],axis=1)
```

```python
df.info()
```

```python
X = df.drop(['county','state','percentage20_Donald_Trump','percentage20_Joe_Biden'],axis=1)
```

```python
y= df[['percentage20_Donald_Trump', 'percentage20_Joe_Biden']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X,
                                                      y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler(with_mean=True)
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense
np.random.seed(7)
#first layer
#Relu maps inputs to 0 or 1
#128 - number of nodes
model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(8, activation='relu'))
#outputlayer
#softmax is a form of logistic regression - turns real values into probabilities
model.add(Dense(2, activation='softmax'))
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=78, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```
